### Cleaning Data

```python
# Menggunakan interpolasi linier untuk kolom polutan karena data cenderung berkesinambungan
prsa_shunyi_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']] = prsa_shunyi_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].interpolate(method='linear')
```

```python
# Menggunakan interpolasi linier untuk kolom cuaca karena data ini juga bersifat kontinu
prsa_shunyi_df[['TEMP', 'PRES', 'DEWP', 'RAIN']] = prsa_shunyi_df[['TEMP', 'PRES', 'DEWP', 'RAIN']].interpolate(method='linear')
```

```python
# Mengisi missing value pada kolom arah angin (wd) menggunakan nilai mode (nilai yang paling sering muncul)
prsa_shunyi_df['wd'] = prsa_shunyi_df['wd'].fillna(prsa_shunyi_df['wd'].mode()[0])
```

```python
# Menggunakan interpolasi linier untuk kolom kecepatan angin (WSPM)
prsa_shunyi_df['WSPM'] = prsa_shunyi_df['WSPM'].interpolate(method='linear')
```

```python
# Mengecek kembali apakah masih ada missing values
print(prsa_shunyi_df.isna().sum())
```

```python
prsa_shunyi_df.describe()
```

**Insight**
- Missing value pada kolom polutan (PM2.5, PM10, SO2, NO2, CO, O3) dan cuaca (TEMP, PRES, DEWP, RAIN, WSPM) sudah diisi menggunakan interpolasi linier.
- Missing value pada kolom arah angin (wd) sudah diimputasi menggunakan nilai mode.
- Semua kolom sudah bebas dari missing values.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada bulan apa rata-rata konsentrasi PM2.5 mencapai nilai tertinggi dalam setahun?

```python
# Mengambil data hanya dari tahun yang relevan, misalnya 2014
prsa_shunyi_df['year'] = prsa_shunyi_df['year'].astype(int)
monthly_data = prsa_shunyi_df[['year', 'month', 'PM2.5']]

# Menghitung rata-rata PM2.5 per bulan
monthly_average = monthly_data.groupby(['year', 'month']).mean().reset_index()

# Mengatur nama bulan
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
monthly_average['month'] = monthly_average['month'].map(lambda x: months[x-1])

# Mengambil tahun tertentu
monthly_average_2014 = monthly_average[monthly_average['year'] == 2014]

# Mem-plot data
plt.figure(figsize=(12, 5))
sns.barplot(x='month', y='PM2.5', data=monthly_average_2014, palette='Blues')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Rata-Rata Konsentrasi PM2.5 (µg/m³)', fontsize=14)
plt.title('Rata-Rata Konsentrasi PM2.5 di Shunyi pada Tahun 2014', fontsize=16)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana hubungan antara suhu udara (TEMP) dan konsentrasi PM2.5 serta PM10?

```python
# Mengambil data untuk analisis
temp_pm_data = prsa_shunyi_df[['TEMP', 'PM2.5', 'PM10']]

# Mem-plot scatter plot untuk PM2.5 dan TEMP
plt.figure(figsize=(12, 5))
sns.scatterplot(x='TEMP', y='PM2.5', data=temp_pm_data, color='blue', label='PM2.5')
sns.scatterplot(x='TEMP', y='PM10', data=temp_pm_data, color='orange', label='PM10')
plt.xlabel('Suhu Udara (°C)', fontsize=14)
plt.ylabel('Konsentrasi (µg/m³)', fontsize=14)
plt.title('Hubungan antara Suhu Udara dan Konsentrasi PM2.5 serta PM10', fontsize=16)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

**Insight**
- Bar chart di atas menunjukkan bahwa bulan Februari 2014 memiliki rata-rata konsentrasi PM2.5 tertinggi, yang mungkin disebabkan oleh peningkatan aktivitas pemanasan selama musim dingin.
- Scatter plot di atas menunjukkan bahwa suhu udara (TEMP) memiliki hubungan negatif dengan konsentrasi PM2.5 dan PM10. Saat suhu meningkat, konsentrasi PM2.5 dan PM10 cenderung menurun, mengindikasikan bahwa kondisi cuaca yang lebih hangat mungkin membantu dalam pengurangan polusi udara.

## Conclusion


- Rata-rata konsentrasi PM2.5 tertinggi di Shunyi terjadi pada bulan Februari 2014.
- Ada hubungan negatif antara suhu udara dan konsentrasi PM2.5 serta PM10, yang menunjukkan bahwa peningkatan suhu dapat berkontribusi pada penurunan konsentrasi polutan.
